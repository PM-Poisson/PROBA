```python
import numpy as np
import matplotlib.pyplot as plt

# liste d'un grand nombre de valeurs aléatoires comprises entre 0 et 1 selon une loi uniforme
N = 100000
X = np.random.rand(N)  # variable aléatoire suivant une loi uniforme entre 0 et 1

print('===== Exemple 1 =====')
plt.subplots(1, 2, figsize=(12, 5))
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

##### histogramme des effectifs de Y
plt.subplot(121)
hX, binsX, _ = plt.hist(X, bins=30, edgecolor='black')
plt.xlabel('valeurs de X')
plt.title('Effectifs')

# informations sur l'histogramme des effectifs
print(f'Nombre de classes : {len(binsX) - 1}')
print(f'Les effectifs : {hX}')
print(f'Somme des effectifs : {np.sum(hX)}')
print('\n')

##### histogramme des probabilités de Y
plt.subplot(122)
hX_prob, binsX_prob, _ = plt.hist(X, bins=30, density=True, edgecolor='black')
plt.xlabel('valeurs de X')
plt.title('Densité de probabilités')

# informations sur l'histogramme des probabilités
print(f'Nombre de classes : {len(binsX_prob) - 1}')
print(f'Les probabilités : {hX_prob * np.diff(binsX_prob)}')
print(f'Somme des probabilités : {np.sum(hX_prob * np.diff(binsX_prob))}')
print('\n')
```
